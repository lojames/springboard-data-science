## DataFrames & Pandas

Pandas is built on NumPy and Matplotlib

Rectangular data = tabular data

Columns have same data type. Each row is a new instance of whatever object.

df.head() shows first few rows of DataFrame

df.info() displays names of columns, datatypes they contain, and if any values are missing

df.shape gives tuple (num rows, num cols)

df.describe() will compute summary statistics (mean, median, standard dev, min , max)

The components of a dataframe:
1) df.values (the row data in the dataframe)
2) df.index (the row index/row names)
3) df.columns (the column names)

## Explicit Indexes
Used mainly because it makes subsetting rows much easier.

Explicit indices are controversion since we are converting data into index values.  This violates "tidy data" principles.


````python
# Subsetting w/o index swapping
subset_no_index = homelessness[homelessness["state"].isin(['Alaska','Arkansas'])]
subset_no_index
````

````python
# Setting a column as the index
homelessness_index = homelessness.set_index('state')
homelessness_index.head()
````

````python
# Accomplishing above subsetting with after index swapping
homelessness_index.loc[['Alaska', 'Arkansas']]
````

````python
# Removing a named index
homelessness_index = homelessness_index.reset_index()
homelessness_index.head()

# args:
#   drop = True/False (will drop the index column from table)
#    
````

````python
# Hierarchical Index - setting multiple indices
homelessness_index = homelessness.set_index(['region','state'])
homelessness_index.head()
````

````python
# Subsetting multiple regions
homelessness_index.loc[['Pacific', 'Mountain']]
````

````python
# Subsetting on regions and state with tuples (much be exact match)
homelessness_index.loc[[('Pacific', 'Alaska'), ('Mountain', 'Colorado')]]
````

````python
# Sorting by index
homelessness_index.sort_index().head()
````

````python
# Sorting by index with args level and ascending
homelessness_index.sort_index(level='state', ascending=False).head()
````

## Slicing and Subsetting with .loc and .iloc

````python
# Review on slicing standard lists

standard_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
standard_list[2:4]
````

````python
# Returning the first n=4 elements in a list
standard_list[:4]
````

````python
# Returning entire list
standard_list[:]
````

**Unlike standard lists, slicing and subsetting with .loc on DataFrames always INCLUDES the last row/column**
**Sort your indices first!**

````python
homelessness_index= homelessness_index.sort_index()
homelessness_index
````

````python
# Slicing by outer index levels is straightforward. When messing with inner levels, always include full tuples.
homelessness_index.loc[('Mountain', 'Idaho'):'New England']
````

````python
# Slicing columns with loc
homelessness_index.loc[:, 'state_pop':'total'].head()
````

````python
# Doing both of above simultaneously with iloc:
homelessness_index.iloc[5:10, 2:5]
````

## Subsetting with Pivot Tables
* The first argument is column name of values to aggregate.
* ***index***, specifies which argument to group by and display in rows.
* ***columns***, specifies which argument to group by and display in columns.
* The default aggregation function is mean.

***Summary statistics, such as mean, have an axis argument.***

***axis***, specifies whether to calculate statistics across rows (index) or columns (columns).  The default argument is index.

## Handling Missing Values
Use the ```.isna()``` method on a DataFrame to get a boolean table of missing values.  More useful is the ```.isna().any()``` which will yield a table of columns and whether any values in the columns are missing. In practice, use ```.isna().sum()``` which will yield a table of columns and the number of missing values.

We can remove the rows with missing values with the ```.dropna()``` method.

We can replace all NaNs with ```.fillna(0)``` *Note that 0 can be anything...such a default string.*


## Creating DataFrames

Can be created by two methods:
1. From a list of dictionary row by row
2. From a dictionary of lists column by column

````python
# Creating a List of Dictionaries

list_of_dicts = [
    {'name':'Ginger', 'breed':'Dachshund', 'height_cm': 22, 'weight_kg': 10, 'date_of_birth': pd.to_datetime('2019-03-14')},
    {'name':'Scout', 'breed':'Dalmation', 'height_cm': 59, 'weight_kg': 25, 'date_of_birth': pd.to_datetime('2019-05-09')}
]
new_dogs=pd.DataFrame(list_of_dicts)
new_dogs
````

````python
# Creating a Dictionaries of Lists

dict_of_lists = {
    'name': ['Ginger', 'Scout'],
    'breed': ['Dachshund', 'Dalmation'],
    'height_cm': [22, 59],
    'weight_kg': [10,25],
    'date_of_birth': pd.to_datetime(['2019-03-14', '2019-05-09'])
}
new_dogs=pd.DataFrame(dict_of_lists)
new_dogs
````
